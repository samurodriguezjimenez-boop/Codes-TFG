```python
import numpy as np

def e0_tfim(g):
    """Ground state energy density of the transverse field Ising model."""
    return -1/(2*np.pi)*np.trapezoid(np.sqrt(1+g**2-2*g*np.cos(np.linspace(-np.pi, np.pi, 1000))), dx=2*np.pi/1000)

def magnetization_tfim(g):
    """Magnetization of the transverse field Ising model."""
    if g < 1:
        return (1 - g**2)**(1/8)
    else:
        return 0
    
def susceptibility_tfim(g):
    """Magnetic susceptibility of the transverse field Ising model."""
    if g < 1:
        return (1 - g**2)**(-7/8)
    else:
        return 0
```

```python
if __name__ == "__main__":
    import matplotlib.pyplot as plt
    J = 1  # Coupling constant
    h_values = np.linspace(0, 3, 100)
    g_values = h_values / J

    e0_values = [e0_tfim(g) for g in g_values]
    mag_values = [magnetization_tfim(g) for g in g_values]
    chi_values = [susceptibility_tfim(g) for g in g_values]

    fig, ax = plt.subplots(1,3, figsize=(18, 5))
    ax[0].plot(g_values, e0_values, label='Energía de estado fundamental $E_0/N$')
    ax[0].set_ylabel('Energía de estado fundamental por sitio, $E_0/N$')

    ax[1].plot(g_values, mag_values, label='Magnetización $M$')
    ax[1].set_ylabel('Magnetización por sitio, $M$')

    ax[2].plot(g_values, chi_values, label='Susceptibilidad $\chi$')
    ax[2].set_ylabel('Susceptibilidad magnética, $\chi$')

    ax[0].set_xlabel('Parámetro de control $g$ $(h / J)$')
    ax[1].set_xlabel('Parámetro de control $g$ $(h / J)$')
    ax[2].set_xlabel('Parámetro de control $g$ $(h / J)$')

    for axis in ax:
        axis.grid()
        axis.legend()
    plt.show()
```
